```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML
```

```python
#Calaculate correlation between every column on the data using Pearson
corr = df.corr('pearson')

#Take the absolute value of correlation
corrs=[abs(corr[attr]['target']) for attr in list(features)]

#make a list of pair[(corr,feature)]
l=list(zip(corrs,list(features)))

#sort the list of pairs in reverse/descendding order
#with the correlation value as the key for sorting
l.sort(key = lambda x: x[0] , reverse = True)

#'Unzip' pair of two list
#zip(*l), take a list that looks like[[a,b,c],[d,e,f],[g,h,i]]
#and returns[[a,d,g],[b,e,h],[c,f,i]]
corrs,labels = list(zip((*l)))


#plot correlation w.r.t the target variable as bar graph
index = np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)

plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variable')

plt.xticks(index,labels)
plt.show()
```

```python
X=df['LSTAT'].values
Y=df['target'].values

#Before normalization
print(Y[:5])
```

```python
x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1,1))
X = X[:, -1]

y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1,1))
Y = Y[:, -1]

#After Normalization
print(Y[:5])
```

```python
def error(m,x,c,t):
    N = x.size
    e = sum(((m * x + c) - t) ** 2)
    return e * 1/(2 * N)
```

```python
#Splitting the dataset into trrain and test datasets
#0.2 indiacates 20% of the data is randomly sampled as testing data
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2)
```

```python
def update(m, x, c, t, learning_rate):
    grad_m = sum(2 * ((m * x + c) - t) * x)
    grad_c = sum(2 * ((m * x + c) - t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m, c
```

```python
def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = list()
    mc_values = list()
    for i in range(iterations):
        e = error(m, x, c, t)
        if e < error_threshold:
            print('Error less than the threshold, Stopping the gradient descent')
            break
        error_values.append(e)
        m, c = update(m, x, c, t, learning_rate)
        mc_values.append((m, c))
    return m, c, error_values, mc_values
    
    
```

```python
%%time
init_m = 0.9
init_c = 0
learning_rate = 0.001
iterations = 250
error_threshold = 0.001

m ,c ,error_values ,mc_values = gradient_descent(init_m, init_c, xtrain, ytrain, learning_rate,iterations, error_threshold)  
```

```python
#visualization of the learning process
plt.scatter(xtrain, ytrain, color='b')
plt.plot(xtrain, (m * xtrain + c), color='r')
```

```python
#plotting error values
plt.plot(np.arange(len(error_values)), error_values)
plt.ylabel('Error')
plt.xlabel('Iterations')
```

```python
#Animation to show how the regression line changes over the course of training
#As the number of iterations increases, changes in the line are less noticable
#In order to reduce processing time for the animation, it is adivasable to choose smaller value
mc_values_anim = mc_values[0:250:5] 
```
