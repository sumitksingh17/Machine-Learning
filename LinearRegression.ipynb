```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML
```

```python
#Load the dataset
boston = load_boston()
#Descrition of the dataset
print(boston.DESCR)
```

```python
#Put the data into pandas DataFrames
features = pd.DataFrame(boston.data, columns = boston.feature_names)
features
```

```python
features['AGE']
```

```python
target = pd.DataFrame(boston.target,columns=['target'])
target
```

```python
max(target['target'])
```

```python
min(target['target'])
```

```python
#Concatenate the features and target into a single DataFrame
#axis=1 makes it concatenate column wise
df = pd.concat([features,target],axis=1)
df
```

```python
#use round(decimals=2) to set the decimal precision
df.describe().round(decimals=2)
```

```python
#Calaculate correlation between every column on the data using Pearson
corr = df.corr('pearson')

#Take the absolute value of correlation
corrs=[abs(corr[attr]['target']) for attr in list(features)]

#make a list of pair[(corr,feature)]
l=list(zip(corrs,list(features)))

#sort the list of pairs in reverse/descendding order
#with the correlation value as the key for sorting
l.sort(key = lambda x: x[0] , reverse = True)

#'Unzip' pair of two list
#zip(*l), take a list that looks like[[a,b,c],[d,e,f],[g,h,i]]
#and returns[[a,d,g],[b,e,h],[c,f,i]]
corrs,labels = list(zip((*l)))


#plot correlation w.r.t the target variable as bar graph
index = np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)

plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variable')

plt.xticks(index,labels)
plt.show()
```
