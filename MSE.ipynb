```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```python
#Using linspace function generate n evelny spaced values from zero to 2PI radians
n = 200
#genetates 200 values
x = np.linspace(0, 2 * np.pi, n)
sin_values = np.sin(x)

#Plot the sin wave
plt.plot(x, sin_values)
```

```python
#Add some noise to the sine wave
noise = 0.5
noisy_sine_values = sin_values + np.random.uniform(-noise, noise, n)

#Plot the noisy values
plt.plot(x, noisy_sine_values, color = 'r')
plt.plot(x, sin_values, linewidth = 3)
```
